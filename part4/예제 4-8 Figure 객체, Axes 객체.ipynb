```python
# Matplotlib 기본 구조 - Figure 객체 vs Axes 객체
```

```python
# Matplotlib에서 그래프를 그리는 과정은 화가가 그림을 그리는 과정에 비유할 수 있음
# 화가는 그림을 그리기 위해 그림틀을 먼저 준비하고, 그 위에 그림을 그림
# Figure객체가 그림틀, Axes 객체가 그위에 그려지는 각각의 그림에 해당함
```

```python
# Figure 객체는 전체 그래프 객체를 담는 하나의 컨테이너를 뜻함, 빈 캔버스
# Axes 객체는 그래프가 그려지는 구체적인 영역을 의미하며, 하나의 Figure 안에 여러가지 Axes 객체를 포함할 수 있다
# 내부에는 축 axis, 그래프 plot, 제목 title 등이 포함됨
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 예제 2.7의 재무제표 API를 활용하여 수집한 데이터
samsung_revenue = pd.read_csv('./data/삼성전자_분기별_매출액.csv')
samsung_revenue = samsung_revenue.sort_values('quarter')
samsung_revenue
```

```python
fig = plt.figure()
print(fig)
# 캔버스의 크기가 640 480, with 0 Axes는 어떠한 그림도 추가되지 않았음을 의미
```

```python
fig, axe = plt.subplots()
print(fig)
print(axe)
# plt.subplots()는 Figure 객체 하나와 그 내부에 Axes 객체 하나를 함께 생성 
# Axes(0.125,0.11;0.775x0.77), Figure 내부에서 Axes가 차지하는 공간을 뜻함
# 0.125, 0.11 은 Axes의 왼쪽 하단 모서리의 위치를 나타내고, 0.775x0.77 은 Axes의 너비와 높이를 비율로 나타냄
```

```python
# 삼성전자의 분기별 매출액 데이터를 Axes 객체에 표시
fig, axe = plt.subplots(figsize = (8,2))
axe.plot(samsung_revenue['quarter'], samsung_revenue['value'])
plt.show()
```

```python
# 1행 2열 구조의 grid에 배치된 2개의 axes 객체와 이를 담고 있는 Figure 객체를 생성 
# 반환된 axes 변수는 2개의 Axes 객체를 포함하는 배열이다 [<Axes: > <Axes: >]
fig, axes = plt.subplots(1,2)
print(fig)
print(axes)
```

```python
fig, axes = plt.subplots(1,2, figsize = (12,2))
axes[0].plot(samsung_revenue['quarter'], samsung_revenue['value'])
samsung_revenue['value'].plot(ax = axes[1])
plt.show()
# 첫 번째 그래프는 'quarter'를 X축으로 사용하지만,
# 두 번째 그래프는 기본적으로 DataFrame의 인덱스를 X축으로 사용함.
# 두 번째 그래프는 인덱스를 X축으로 사용하기 때문에, 횟수(데이터의 순서) 기준으로 배치돼서 나옴.
```

```python
# 첫 번째 그래프(axes[0])
# X축: 'quarter' (2022-Q2 ~ 2023-Q3)
# Y축: 'value' (매출)
# 분기별 매출 변화를 선 그래프로 표현 (분기 값이 X축에 정확히 매칭됨)

# 두 번째 그래프(axes[1])
# X축: 인덱스 (0,1,2,3,4,5)
# Y축: 'value'
# X축에 quarter 대신 인덱스(0~5)가 배치된 매출 변화 그래프가 됨.
```

```python
fig, axes = plt.subplots(2,2)
print(fig)
print(axes)
```

```python
# axe 객체에 차트 추가
fig, axes = plt.subplots(2,2, figsize = (12,2), constrained_layout = True)
axes [0,1].plot(samsung_revenue['quarter'], samsung_revenue['value'])
plt.show()

# constrained_layout = True 간격을 자동으로 조절하여 축 레이블, 타이틀 등이 서로 겹치지 않게 하는 설정
```

```python
# plt.subplot_mosaic 입력된 배열구조에 다라 Axes 객체들을 배치함
fig, axes = plt.subplot_mosaic([['top_left','right'],
                                ['bottom_left', 'right']])

# top_left bottom_left Figure 내부의 왼쪽에 2개의 Axes 객체를 생성
# right Figure 내부의 오른쪽에 하나의 Axes를 두 번 반복하여 같은 위치를 참조하여 공유
print(fig)
print(axes)
```

```python
fig, axes = plt.subplot_mosaic([['top_left','right'],
                                ['bottom_left', 'right']],
                                   figsize = (12,4))
axes['right'].plot(samsung_revenue['quarter'], samsung_revenue['value'])
plt.show()
```

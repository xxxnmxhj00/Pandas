```python
# jointplot() 함수는 두 변수 간의 관계를 중앙에 크게 기본 차트로 표시하고
# x - y 축의 마진 영역에 각 변수의 분포를 보여주는 히스토그램 또는 KDE 플롯을 동시에 보여준다
# 두 변수의 관계와 각 변수의 분포를 한눈에 파악하기 좋음
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

sns.set_style('whitegrid')

# 조인트 그래프 - 산점도(기본값)
j1 = sns.jointplot(x = 'fare', y = 'age', data = titanic)

# 조인트 그래프 - 회귀선 kind = 'reg' -> 중앙에 선형 회귀선을 포함한 산점도를 그림
j2 = sns.jointplot(x = 'fare', y = 'age', kind = 'reg', data = titanic)

# 조인트 그래프 - 육각 그래프 kind = 'hex' -> 육각형을 사용하여 데이터의 밀도를 나타내는 육각 그래프 표시
j3 = sns.jointplot(x = 'fare', y = 'age', kind = 'hex', data = titanic)

# 조인트 그래프 - 커널 밀도 그래프 kind = 'kde' -> 두 변수간의 밀도를 표현하는데 커널 밀도 추정을 사용
# 두 변수의 연속적인 부드러운 곡선으로 보여줌 , 오른쪽과 위쪽의 마진 플롯에도 커널 밀도 그래프가 표시됨
j4 = sns.jointplot(x = 'fare', y = 'age', kind = 'kde', data = titanic)


plt.show()
```

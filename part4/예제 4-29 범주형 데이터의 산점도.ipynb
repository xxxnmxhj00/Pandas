```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

# 스타일 테마 설정 (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('whitegrid')

# 그래프 객체 생성(figure에 2개의 서브 플롯을 생성)
fix, axes = plt.subplots(1,2, figsize = (15,5))

# stripplot() 함수는 카테고리별로 데이터를 점으로 표현하여 분포를 나타냄
# 각 데이터 포인트들이 겹칠 수 있어 데이터의 밀도를 정확히 표현하는데 한계가 있음
# 이산형 변수의 분포 - 데이터 분산 미고려
sns.stripplot(x = 'class',
              y = 'age',
              data = titanic,
              ax = axes[0]) # Axes 객체 = 첫 번째 그래프

# 이산형 변수의 분포 - 데이터 분산 고려(중복 x)
# swarmplot() 함수는  데이터의 분산까지 고려하여, 데이터 포인트가 서로 겹치지 않도록 조정하여 표시
sns.swarmplot(x = 'class',
              y = 'age',
              data = titanic,
              ax = axes[1], # Axes 객체 = 두 번째 그래프
              hue = 'class', # 색상 구분, 클래스별로 다른 색상 자동 적용 
              size = 4) # 마커 크기

# 차트 제목 표시
axes[0].set_title('Strip Plot')
axes[1].set_title('Swarm Plot')

plt.show()
```

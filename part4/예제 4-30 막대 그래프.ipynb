```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

# 스타일 테마 설정 (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('ticks')

# 그래프 객체 생성(figure에 2개의 서브 플롯을 생성)
fix, axes = plt.subplots(2,2, figsize = (10,6), constrained_layout = True)

# stripplot() 함수는 카테고리별로 데이터를 점으로 표현하여 분포를 나타냄
# 각 데이터 포인트들이 겹칠 수 있어 데이터의 밀도를 정확히 표현하는데 한계가 있음
# 이산형 변수의 분포 - 데이터 분산 미고려
sns.barplot(x = 'sex',
              y = 'survived',
              data = titanic,
              ax = axes[0,0]) # Axes 객체 = 첫 번째 그래프
# 기본적으로 y 축 변수의 평균을 계산하며, 검정색 오차 막대는 평균 모수추정의 신뢰 구간을 나타낸다
# 이와 같은 통계 연산이 Seaborn 라이브러리의 특징

# 이산형 변수의 분포 - 데이터 분산 고려(중복 x)
# swarmplot() 함수는  데이터의 분산까지 고려하여, 데이터 포인트가 서로 겹치지 않도록 조정하여 표시
sns.barplot(x = 'sex',
              y = 'survived',
              data = titanic,
              errorbar = None, # 오차 막대를 표시하지 않는다, 막대의 높이만 사용하여 생존률을 비교하는데 유용
              ax = axes[0,1]) # 마커 크기

sns.barplot(x = 'sex',
              y = 'survived',
              hue = 'class',
              data = titanic,
              ax = axes[1,0], # 마커 크기
errorbar = ('ci', 95), estimator='median')  # 생존률 중앙값 계산, 95% 신뢰구간을 오차 막대로 표시


sns.barplot(x = 'sex',
              y = 'survived',
              hue = 'class',
              dodge = False, # 각 객실 등급별 막대가 서로 겹쳐서 표시
              data = titanic,
              ax = axes[1,1]) # 마커 크기

# 차트 제목 표시
axes[0,0].set_title('titanic survived - sex')
axes[0,1].set_title('titanic survived - sex(w/o errorbar)')
axes[1,0].set_title('titanic survived - sex/class(median)')
axes[1,1].set_title('titanic survived - sex/class(dodge)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

# 스타일 테마 설정 (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')

# 그래프 객체 생성(figure에 2개의 서브 플롯을 생성)
fix, axes = plt.subplots(1,2, figsize = (15,5))

# 그래프 그리기 - 선형회귀선 표시(fig_reg = True)
# regplot 함수는 서로 다른 2개의 연속 변수 사이의 산점도를 그리고 선형회귀분석에 의한 회귀선을 함께 나타낸다
# fit_reg = False 옵션을 설정하면 회귀선이 안보이게 할 수 있다
sns.regplot(x = 'age',
            y = 'fare',
            data = titanic,
            ax = axes[0]) # Axes 객체 = 첫 번째 그래프
sns.regplot(x = 'age',
            y = 'fare',
            data = titanic,
            ax = axes[1], # Axes 객체 = 두 번째 그래프
            fit_reg = False) # 회귀선 미표시

plt.show()
```

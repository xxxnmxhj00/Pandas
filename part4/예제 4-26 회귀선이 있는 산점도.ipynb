```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

#    (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')

#   (figure 2   )
fix, axes = plt.subplots(1,2, figsize = (15,5))

#   -  (fig_reg = True)
# regplot    2          
# fit_reg = False       
sns.regplot(x = 'age',
            y = 'fare',
            data = titanic,
            ax = axes[0]) # Axes  =   
sns.regplot(x = 'age',
            y = 'fare',
            data = titanic,
            ax = axes[1], # Axes  =   
            fit_reg = False) #  

plt.show()
```

```python
# 히스토그램은 변수가 하나인 단변수 데이터의 빈도수를 그래프로 표현
# x 축을 같은 크기의 여러 구간으로 나누고 각 구간에 속하는 데이터 값의 개수(빈도)를 y축에 표시
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('classic')

df = pd.read_csv('./data/auto-mpg.csv', header = None)

# 열 이름 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'accerleration','model year','origin','name']

# 연비 열에 대한 히스토그램 그리기
df['mpg'].plot(kind = 'hist', bins = 10, color = 'coral', figsize = (10,5))

plt.title('Histogram')
plt.xlabel('mpg')
plt.show()
```

```python
# 원산지 국가 origin 열을 기준으로 그룹별로 히스토그램 그리기
df[['mpg', 'origin']].plot(by = ['origin'], kind = 'hist', figsize = (8,10));
```

```python
# 박스플롯과 바이올린 그래프는 데이터의 분포와 중심 경향, 이상치 등을 시각적으로 탐색하는데 유용
# 박스 플롯은 범주형 데이터 분포와 주요 통계 지표를 함께 제공
# 다만 박스 플롯만으로는 데이터가 퍼져 있는 분산 정도를 정확하게 알기 어렵기 때문에
# 커널밀도함수 그래프를 y축 방향에 추가하여 바이올린 그래프를 그리는 겨웅도 있음
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

sns.set_style('whitegrid')

fix, axes = plt.subplots(2,2, figsize = (15,10))

# 기본값
# 각 범주에 속하는 데이터의 개수를 빈도를 막대의 높이로 나타내는 countplot() 함수
# 박스 그래프 - 기본 값, 데이터의 중앙값, 사분위수, 이상치 등을 시각적으로 표현
sns.boxplot(x = 'alive', y = 'age', data = titanic, ax = axes [0,0])

# 박스 그래프 - hue 변수 추가, 성별 변수를 추가하여, 생존 여부와 성별에 따른 age 분포를 추가하여 비교 
sns.boxplot(x = 'alive', y = 'age', hue = 'sex', data = titanic, ax = axes [0,1])

# 바이올린 그래프 - 기본 값, 박스 그래프의 정보에 커널 밀도 추정을 추가하여 age의 분포가 생존 여부에 따라서 어떻게 다른지 연속적으로 보여줌
sns.violinplot(x = 'alive', y = 'age', data = titanic, ax = axes [1,0])

# 바이올린 그래프 - hue 변수 추가
sns.violinplot(x = 'alive', y = 'age', hue = 'sex', data = titanic, ax = axes [1,1])


plt.show()
```

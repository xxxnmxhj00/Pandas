```python
# 첫 행에 데이터프레임 df의 클래스 유형인 'pandas.core.frame.DataFrame' 출력
# 398개의 행 인덱스와 9개의 열에 관한 정보
# 각 열의 이름과 데이터 개수(Non - Null : 누락 데이터가 아닌 유효한), 자료형이 출력
# 자료형(실수형 4개, 정수형 3개, 문자열 2개)과 메모리 사용량(28.1 + KB) 표시
```

```python
# info() 메소드를 사용하지 않고 데이터프레임 클래스의 dtypes 속성을 활용하여 각 열의 자료형 확인 가능
df.dtypes
```

```python
# 특정 열을 선택하면 시리즈 객체가 추출되는데, 시리즈 객체의 dtypes 속성을 활용하면 해당 열을 구성하는 데이터 자료형 확인 가능
df['mpg'].dtypes
```

```python
# 데이터프레임에 describe() 메소드를 적용하면 ,산술 데이터를 갖는 열에 대한 주요 기술 통계 정보
# (평균 표준편차 최댓값 최솟값 중앙값)등을 요약하여 출력
df.describe()
```

```python
# 문자열 등 산술 데이터가 아닌 열에 대한 정보를 포함하고 싶을 때는 include = 'all'
df.describe(include = 'all')
# name 열의 unique(고유값 개수), top(최빈값), freq(빈도수) 정보 추가
```

```python
# 숫자형 데이터를 가진 열을 선택하여 통계적 요약을 보여줌
df.describe(include = 'number')
```

```python
df.describe(include = 'object')
```

```python
# 특정 자료형을 포함하는 열만을 선택
df.describe(include = ['number','object'])
```

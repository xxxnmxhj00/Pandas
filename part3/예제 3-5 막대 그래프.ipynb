```python
# plot() 메소드로 선 그래프가 아닌 다른 종류의 그래프를 그리려면 kind 옵션에 그래프 종류를 지정한다
# 막대 그래프를 그리려면 다음과 같이 kind = 'bar' 옵션을 추가한다
```

```python
# 막대 그래프 : DataFrame 객체.plot(kind = 'bar')
```

```python
import pandas as pd

df = pd.read_excel('./data/남북한발전전력량.xlsx')

df_ns = df.iloc[[0,5], 3:]

df_ns.index = ['South','North']

# 발전량 데어터의 자료형을 문자형에서 정수형으로 변환 
df_ns = df_ns.astype(int)

# 행, 열 전치하여 막대 그래프 그리기
tdf_ns = df_ns.T
print(tdf_ns.head())
print('\n')
tdf_ns.plot(kind='bar');
```

```python
tdf_ns.plot(kind = 'barh')
```

```python
tdf_ns.plot(kind='bar', stacked = True);
# stacked = True 옵션을 적용하면 x축에 순서대로 배치된 데이터프레임 행에 대해서 각 열에 해당하는 값들을 막대 위에 쌓아서 표시
```

```python
import pandas as pd

df = pd.read_csv('./data/auto-mpg.csv', header = None)

# 열 이름을 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'accerleration','model year','origin','name']

# 열을 선택하여 박스 플롯 그리기
df [['mpg','cylinders']].plot(kind = 'box');
```

```python
# 박스 플롯은 특정 변수의 데이터 분포와 분산 정도에 대한 정보를 제공한다
# 연비는 10 ~ 45 범위에 넓게 분포 되어 있고 'o' 표시의 이상값도 확인된다
# 반면 실린더 게수는 좁은 범위에 몰려 있다
# 이처럼 각 변수들의 데이터가 퍼져 있는 정도를 확인할 때 쓰인다
```
